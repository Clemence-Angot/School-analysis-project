```python
#importation of traditional librairies

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import seaborn as sns
```

```python
#open my CSV

monFichier = pd.read_csv('datas.csv')
france_population = pd.read_csv('population_of_france.csv')
```

```python
#print my original dataset

monFichier
```

```python
#print my column's names

monFichier.columns
```

```python
#create a column for the sum

monFichier['Sum'] = None
monFichier.to_csv('datas.csv', index = False)
```

```python
#sum of the number for all lines (usefull for the mortality)

columns_list = list(monFichier)
columns_list.remove('Year')
columns_list.remove('Sex')
columns_list.remove('Type')
columns_list.remove('Cancer_Type')
monFichier['Sum'] = monFichier[columns_list].sum(axis = 1)
```

# Uterine cancer data analysis

```python
#create a new dataframe with only Uterine cancer datas

uterine_df = monFichier[(monFichier['Cancer_Type'] == 'Uterine cancer')]

#print my dataset with uterine cancer

uterine_df
```

```python
#split the dataset in two : mortality and incidence and del the lines without values

uterine_df_mortality = uterine_df[(uterine_df['Type'] == 'Mortality')].dropna()
uterine_df_incidence = uterine_df[(uterine_df['Type'] == 'Incidence')].dropna()
```

```python
#print my datas : mortality of uterine cancer

uterine_df_mortality
```

```python
#print my datas : incidence of uterine cancer

uterine_df_incidence
```

```python
#Creation of a graph that will show the evolution of the number of deaths from Uterine cancer between 1968 and 2010.
#Using the plot function

x = uterine_df_mortality['Year']
y = uterine_df_mortality['Sum']

plt.plot(x, y, color = 'purple')

plt.title("Evolution of the number of deaths from uterine cancer between 1968 and 2011\n")
plt.xlabel("Years")
plt.ylabel("Nomber of deaths")

plt.show
```

```python
#Creation of a graph that will show the evolution of the number of people diagnosed with Uterine cancer between 1968 and 2011.
#Using the plot function

x = uterine_df_incidence['Year']
y = uterine_df_incidence['Sum']

plt.plot(x, y, color = 'green')

plt.title("Evolution of the number of incidence from uterine cancer between 1968 and 2011\n")
plt.xlabel("Years")
plt.ylabel("Nomber of cases")

plt.show
```

```python
#Creation of a graph that will show the evolution of the case fatality rate of the Uterine cancer.

one = list(uterine_df_mortality['Sum'])[14:43]
two = list(uterine_df_incidence['Sum'])
pourcentage_of_deaths = []

for index in range(len(one)):
    try: 
        pourcentage_of_deaths.append(float(one[index]/two[index]))
    except:
        pourcentage_of_deaths.append(None)

x = uterine_df_incidence['Year']
y = pourcentage_of_deaths

plt.plot(x, y, color = 'red')

plt.title("Evolution of the case fatality rate from uterine cancer between 1982 and 2010\n")
plt.xlabel("Years")
plt.ylabel("Case fatality rate")

plt.show
```

```python
#Representation of the mortality rate

population = list(map(lambda x : int(x), list(france_population['Female Population'])))
mortality_rate = []

for idx in range(len(one)):
    try: 
        mortality_rate.append(float(one[idx]/population[idx]))
    except:
        mortality_rate.append(None)

x = uterine_df_incidence['Year']
y = mortality_rate

plt.plot(x, y, color = 'pink')

plt.title("Evolution of the mortality rate from uterine cancer between 1982 and 2010 \n")
plt.xlabel("Years")
plt.ylabel("Mortality Rate")

plt.show
```

```python
#representation of the prevalence

nbr_sick_people = list(uterine_df_incidence['Sum'])
prevalence = []

for i in range(len(nbr_sick_people)):
    try:
        prevalence.append(float(nbr_sick_people[i])/population[i])
    except:
        prevalence.append(None)

x = uterine_df_incidence['Year']
y = prevalence

plt.plot(x, y, color = 'brown')

plt.title("Evolution of the prevalence from uterine cancer between 1982 and 2010 \n")
plt.xlabel("Years")
plt.ylabel("Prevalence")
```
